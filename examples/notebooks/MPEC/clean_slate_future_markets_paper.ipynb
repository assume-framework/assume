```python
# read storages similar to pp_units
storage_units = pd.read_csv(
    os.path.join(inputs_dir, scenario, "storage_units.csv"), index_col=0
)

# fill addtional_cost_charge and additional_cost_cischarge with 1 to allow for proper depiction of k_values for storages without cost
storage_units["additional_cost_charge"]=storage_units["additional_cost_charge"].fillna(1.0)
storage_units["additional_cost_discharge"]=storage_units["additional_cost_discharge"].fillna(1.0)

storage_units.head()
```

```python
# --- 1) storage_units vorbereiten (Index = unit_id) ---
storage_costs = storage_units.copy()


# Robuste Spaltennamen (je nach CSV/Schema)
charge_col = (
    "additional_cost_charge"
    if "additional_cost_charge" in storage_costs.columns
    else "additional_costs_charge"
)
discharge_col = (
    "additional_cost_discharge"
    if "additional_cost_discharge" in storage_costs.columns
    else "additional_costs_discharge"
)

# --- 2) Mapping für Storage-Kosten ---
charge_map = storage_costs[charge_col].to_dict()
discharge_map = storage_costs[discharge_col].to_dict()

# Nur Rows, die zu Storage-Units gehören
is_storage = market_orders_df["unit_id"].isin(storage_costs.index)

# --- 3) Marginal Cost für Storage aus Richtung (volume) setzen ---
storage_mc = np.where(
    market_orders_df["volume"] < 0,
    market_orders_df["unit_id"].map(charge_map),
    np.where(
        market_orders_df["volume"] > 0,
        market_orders_df["unit_id"].map(discharge_map),
        (market_orders_df["unit_id"].map(charge_map) + market_orders_df["unit_id"].map(discharge_map))
        / 2,
    ),
)

# Nur für Storage-Zeilen schreiben
market_orders_df.loc[is_storage, "marginal_cost"] = storage_mc[is_storage]

# --- 4) k_df aufteilen: Generatoren vs. Storage ---
storage_names = set(storage_costs.index)

# Storage: getrennte k-Werte für Charge und Discharge
storage_rows = market_orders_df[market_orders_df["unit_id"].isin(storage_names)].copy()

# Charge-Gebote: volume < 0 → Speicher will laden
storage_charge_rows = storage_rows[storage_rows["volume"] < 0].copy()
storage_charge_rows["k"] = (
    storage_charge_rows["price"]
    / storage_charge_rows["unit_id"].map(charge_map)
)
storage_charge_rows["k"] = (
    storage_charge_rows["k"].replace([np.inf, -np.inf], 1.0).fillna(1.0)
)

# Discharge-Gebote: volume > 0 → Speicher will entladen
storage_discharge_rows = storage_rows[storage_rows["volume"] > 0].copy()
storage_discharge_rows["k"] = (
    storage_discharge_rows["price"]
    / storage_discharge_rows["unit_id"].map(discharge_map)
)
storage_discharge_rows["k"] = (
    storage_discharge_rows["k"].replace([np.inf, -np.inf], 1.0).fillna(1.0)
)


print(f"Storage charge bids: {len(storage_charge_rows)} rows")
print(f"Storage discharge bids: {len(storage_discharge_rows)} rows")
# Prüfen ob es Stunden gibt wo ein Storage weder lädt noch entlädt:
n_total = len(storage_rows)
n_zero = len(storage_rows[storage_rows["volume"] == 0])
print(f"Storage rows with volume=0 (kein Gebot): {n_zero}/{n_total}")
```
