```python
import pandas as pd

# ---- Load all relevant data ----
plants = pd.read_csv("powerplant_units.csv")
buses = pd.read_csv("buses.csv")
avail = pd.read_csv("availability_df.csv")  # only header check

# ---- Check 1: Availability columns vs plant names ----
avail_cols = set(avail.columns) - {"datetime", "time", "snapshot"}
plants_set = set(plants["name"])
missing_in_plants = avail_cols - plants_set
print("Availability columns not in plants:", len(missing_in_plants), list(missing_in_plants)[:10])

# ---- Check 2: All buses used by plants exist ----
missing_buses = set(plants["node"]) - set(buses["name"])
print("Missing buses:", len(missing_buses), missing_buses)

# ---- Optional: sanity on numeric values ----
sample = pd.read_csv("availability_df.csv")
if sample.iloc[:,1:].isna().any().any():
    print("⚠️ Some NaNs in availability values")
else:
    print("✅ No NaNs in availability values")
if (sample.iloc[:,1:] < 0).any().any() or (sample.iloc[:,1:] > 1).any().any():
    print("⚠️ Some availability values outside [0,1]")
else:
    print("✅ All availability values within [0,1]")
```
